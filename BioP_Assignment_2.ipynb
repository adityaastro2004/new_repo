```python
%%bash
cat > ions.mdp << 'EOF'
integrator  = steep
nsteps      = 2000
emtol       = 1000.0
cutoff-scheme = Verlet
coulombtype = PME
vdwtype     = cut-off
rcoulomb    = 1.0
rvdw        = 1.0
pbc         = xyz
EOF
```

```python
# preparing for ion addition with grompp
# genrate .tpr file using ions.mdp file
!gmx grompp -f ions.mdp -c DQ2.5_glia_gama_4a_ef_SOLV.gro -p topol.top -o ions.tpr -maxwarn 1
```

```python
#adding ions with genion
#use .tpr file to neutralize the system and we generally use NA(Sodium) and CL (Chlorine) for neutralization
# select group: 13 (SOL) to replace the water to add ions
!gmx genion -s ions.tpr -o DQ2.5_glia_gama_4a_ef_SOLV_IONS.gro -p topol.top -pname NA -nname CL -neutral #pname(positive charge name) and nname(negatice charge name)
```

```python
%%bash
cat > minim.mdp << 'EOF'
integrator  = steep
nsteps      = 50000
emtol       = 1000.0
cutoff-scheme = Verlet
coulombtype = PME
vdwtype     = cut-off
rcoulomb    = 1.0
rvdw        = 1.0
pbc         = xyz
constraints = none
EOF
```

```python
# Enegry mimimization to stablize the molecule
#generate .tpr file using minim.mdp files
!gmx grompp -f minim.mdp -c DQ2.5_glia_gama_4a_ef_SOLV_IONS.gro -p topol.top -o em.tpr #em(energy minimization)
```

```python
# run energy minimization using the em.tpr files
#for running gmx mdrun, we always use prefix of .tpr file. So all the output file have same prefix generated by running energy minimization.
!gmx mdrun -v -deffnm em # here deffnm(define file name) and em is used to generate output file with that name
```

```python
# visualize em.gro file
import py3Dmol

# Load the .gro file
with open('em.gro', 'r') as f:
    gro_data = f.read()

# Initialize 3D viewer
view = py3Dmol.view(width=800, height=600)

# Load GRO structure
view.addModel(gro_data, 'gro')

# --- Visualization styles ---
# Protein as cartoon (colored by spectrum)
#view.setStyle({'protein': True}, {'cartoon': {'color': 'spectrum', 'radius': 1.0}})
view.setStyle({'cartoon': {'color': 'spectrum'}})
# Water molecules (resname SOL) as sticks
view.addStyle({'resn': 'SOL'}, {'stick': {'colorscheme': 'cyanCarbon', 'radius': 0.15}})

# Sodium ions (resname NA) as spheres
view.addStyle({'resn': 'NA'}, {'sphere': {'color': 'blue', 'radius': 1.5}})

# Chloride ions (resname CL) as spheres
view.addStyle({'resn': 'CL'}, {'sphere': {'color': 'red', 'radius': 1.5}})
# big sphere molecule represents the IONS for knowing what is ions you can check topol.top file(bottom lines)
# At bottom of topol.top file you will information of number of water and ions molecules, types of forcefields and water model
# Zoom and show
view.zoomTo()
view.show()
```

```python
# to ensure the energy minimization was succesful, we want visualize the energy plot
!gmx energy -f em.edr -o potential.xvg # go to the cursor below and you need to give prompt 10(for potential energy) and then 0, finally click enetr to get the output
```

```python
import matplotlib.pyplot as plt

# --- CONFIG ---
xvg_file = "potential.xvg"

# --- Read .xvg file ---
x = []
y = []

with open(xvg_file, 'r') as f:
    for line in f:
        line = line.strip()
        if line.startswith(('#', '@')) or not line:
            # Skip comments and metadata lines
            continue
        parts = line.split()
        if len(parts) >= 2:
            x.append(float(parts[0]))
            y.append(float(parts[1]))

# --- Plot ---
plt.figure(figsize=(8,5))
plt.plot(x, y, label="Potential Energy", color='blue', linewidth=1.5)
plt.xlabel("Step or Time (ps)")
plt.ylabel("Potential Energy (kJ/mol)")
plt.title("Potential Energy during Energy Minimization")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
#check temp stability
#Command
!gmx energy -f nvt.edr -o temperature.xvg
```

```python
import numpy as np
import matplotlib.pyplot as plt

def load_xvg(path):
    data = []
    with open(path) as f:
        for line in f:
            if line.startswith(('#','@')):
                continue
            data.append([float(x) for x in line.split()])
    return np.array(data)

temp = load_xvg('temperature.xvg')
plt.plot(temp[:,0], temp[:,1])
plt.xlabel('Time (ps)')
plt.ylabel('Temperature (K)')
plt.title('NVT: Temperature vs Time')
plt.savefig('temp_nvt.png', dpi=300)
plt.show()
```

```python
%%bash
cat > npt.mdp << 'EOF'
title       = NPT
integrator  = md
dt          = 0.002
nsteps      = 50000      ; 100 ps
nstxout     = 500
nstvout     = 500
nstenergy   = 500
nstlog      = 500
continuation    = yes
constraint_algorithm = lincs
constraints = h-bonds
cutoff-scheme = Verlet
coulombtype = PME
vdwtype     = cut-off
rcoulomb    = 1.0
rvdw        = 1.0
tcoupl      = V-rescale
tc-grps     = Protein Water_and_ions
tau-t       = 0.1 0.1
ref-t       = 312 312
pcoupl      = Parrinello-Rahman
pcoupltype  = isotropic
tau-p       = 2.0
ref-p       = 1.0
compressibility = 4.5e-5
pbc         = xyz
EOF
```

```python
# pressure and density stability
!gmx energy -f npt.edr -o pressure.xvg
!gmx energy -f npt.edr -o density.xvg
```

```python
import numpy as np
import matplotlib.pyplot as plt

def load_xvg(path):
    data = []
    with open(path) as f:
        for line in f:
            if line.startswith(('#','@')):
                continue
            data.append([float(x) for x in line.split()])
    return np.array(data)

pressure_ = load_xvg('pressure.xvg')
density_ = load_xvg('density.xvg')
t_press=pressure_[:,0]
P=pressure_[:,1]
t_dens=density_[:,0]
rho=density_[:,1]

plt.figure()
plt.plot(t_press, P)
plt.xlabel('Time (ps)')
plt.ylabel('Pressure (bar)')
plt.title('NPT: Pressure vs Time')
plt.savefig('pressure_npt.png', dpi=300)
plt.show()

plt.figure()
plt.plot(t_dens, rho)
plt.xlabel('Time (ps)')
plt.ylabel('Density (kg/m^3)')
plt.title('NPT: Density vs Time')
plt.savefig('density_npt.png', dpi=300)
plt.show()
```
